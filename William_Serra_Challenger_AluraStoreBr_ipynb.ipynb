## 1. O Que é a Média?

-   A **média** representa o valor central do faturamento por compra em uma determinada loja.
-   Quanto maior a média, maior o valor típico gasto pelos clientes.

**Exemplo:**

-   **Loja 1** : Média de R$ 650,49

-   **Loja 2** : Média de R$ 630,97

-   **Loja 3** : Média de R$ 620,61

-   **Loja 4** : Média de R$ 587,15

----------

## 2. O Que é o Desvio Padrão?

-   O desvio padrão mede a dispersão dos valores em torno da média.
-   Um **desvio padrao alto** indica que há grande variação nos valores de compra (compras muito caras ou muito baratas).
-   Um **desvio padrao baixo** sugere que a maioria das compras está próxima da média.

**Exemplo:**

-   **Loja 1** : Desvio padrao de R$ 819,58

-   **Loja 2** : Desvio padrao de R$ 794,08

-   **Loja 3** : Desvio padrao de R$ 774,19

-   **Loja 4** : Desvio padrao de R$ 758,34

----------

## 3. Interpretação dos Resultados

### **Loja 4**

-   Indica que a loja tem um perfil mais uniforme de vendas , com valores de compra concentrados em torno da média.

----------

## 4. Conclusão

# 2. Vendas por Categoria


```python
# Função para contar vendas por categoria
def contar_vendas_por_categoria(df):
    # Contar o número de vendas por categoria
    vendas_por_categoria = df['Categoria do Produto'].value_counts().reset_index()
    vendas_por_categoria.columns = ['Categoria', 'Número de Vendas']
    return vendas_por_categoria
```

```python
loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)
```

```python
# Aplicar a função para cada loja
vendas_categoria_loja1 = contar_vendas_por_categoria(loja)
vendas_categoria_loja2 = contar_vendas_por_categoria(loja2)
vendas_categoria_loja3 = contar_vendas_por_categoria(loja3)
vendas_categoria_loja4 = contar_vendas_por_categoria(loja4)

# Exibir resultados
print("Loja 1:")
print(vendas_categoria_loja1.to_string(index=False))
print("\nLoja 2:")
print(vendas_categoria_loja2.to_string(index=False))
print("\nLoja 3:")
print(vendas_categoria_loja3.to_string(index=False))
print("\nLoja 4:")
print(vendas_categoria_loja4.to_string(index=False))
```

# 2.1 Cálculo do Ticket Médio


```python
# Calcular o Ticket Médio Geral por Loja
# Função para calcular o ticket médio geral por loja
def calcular_ticket_medio_geral(df):
    faturamento_total = df['Preço'].sum()
    numero_de_vendas = df.shape[0]
    ticket_medio_geral = faturamento_total / numero_de_vendas
    return ticket_medio_geral

# Calcular o ticket médio geral para cada loja
ticket_medio_geral_loja1 = calcular_ticket_medio_geral(loja)
ticket_medio_geral_loja2 = calcular_ticket_medio_geral(loja2)
ticket_medio_geral_loja3 = calcular_ticket_medio_geral(loja3)
ticket_medio_geral_loja4 = calcular_ticket_medio_geral(loja4)

# Exibir resultados
print(f"Ticket Médio Geral - Loja 1: R${ticket_medio_geral_loja1:.2f}")
print(f"Ticket Médio Geral - Loja 2: R${ticket_medio_geral_loja2:.2f}")
print(f"Ticket Médio Geral - Loja 3: R${ticket_medio_geral_loja3:.2f}")
print(f"Ticket Médio Geral - Loja 4: R${ticket_medio_geral_loja4:.2f}")
```

```python
import matplotlib.pyplot as plt

# Dados para o gráfico
lojas = ["Loja 1", "Loja 2", "Loja 3", "Loja 4"]
tickets_medios = [ticket_medio_geral_loja1, ticket_medio_geral_loja2, ticket_medio_geral_loja3, ticket_medio_geral_loja4]

# Criar o gráfico de setores
plt.figure(figsize=(8, 6))
plt.pie(tickets_medios, labels=lojas, autopct='%1.1f%%', startangle=90, colors=['blue', 'green', 'orange', 'red'])

# Adicionar título
plt.title("Ticket Médio Geral por Loja")

# Exibir o gráfico
plt.tight_layout()
plt.show()
```

```python
#Calcular o Ticket Médio por Categoria
# Função para calcular o ticket médio por categoria
def calcular_ticket_medio_por_categoria(df):
    # Agrupar por categoria e calcular o faturamento total e número de vendas
    ticket_medio = df.groupby('Categoria do Produto').agg(
        Faturamento_Total=('Preço', 'sum'),
        Numero_de_Vendas=('Preço', 'size')
    ).reset_index()

    # Calcular o ticket médio
    ticket_medio['Ticket Médio'] = ticket_medio['Faturamento_Total'] / ticket_medio['Numero_de_Vendas']

    return ticket_medio

# Aplicar a função para todas as lojas
ticket_medio_loja = calcular_ticket_medio_por_categoria(loja)
ticket_medio_loja2 = calcular_ticket_medio_por_categoria(loja2)
ticket_medio_loja3 = calcular_ticket_medio_por_categoria(loja3)
ticket_medio_loja4 = calcular_ticket_medio_por_categoria(loja4)

# Exibir resultados
print("Loja 1 - Ticket Médio por Categoria:")
print(ticket_medio_loja)
print("\nLoja 2 - Ticket Médio por Categoria:")
print(ticket_medio_loja2)
print("\nLoja 3 - Ticket Médio por Categoria:")
print(ticket_medio_loja3)
print("\nLoja 4 - Ticket Médio por Categoria:")
print(ticket_medio_loja4)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Extrair categorias únicas (usando Loja 1 como referência)
categorias = ticket_medio_loja['Categoria do Produto']

# Ticket médio por categoria para cada loja
ticket_medio_loja1 = ticket_medio_loja['Ticket Médio']
ticket_medio_loja2 = ticket_medio_loja2['Ticket Médio']
ticket_medio_loja3 = ticket_medio_loja3['Ticket Médio']
ticket_medio_loja4 = ticket_medio_loja4['Ticket Médio']

# Configurações do gráfico
x = np.arange(len(categorias))  # Posições no eixo X
width = 0.2  # Largura das barras

fig, ax = plt.subplots(figsize=(12, 6))

# Criar barras para cada loja
bar1 = ax.bar(x - 1.5 * width, ticket_medio_loja1, width, label='Loja 1', color='blue')
bar2 = ax.bar(x - 0.5 * width, ticket_medio_loja2, width, label='Loja 2', color='green')
bar3 = ax.bar(x + 0.5 * width, ticket_medio_loja3, width, label='Loja 3', color='orange')
bar4 = ax.bar(x + 1.5 * width, ticket_medio_loja4, width, label='Loja 4', color='red')

# Adicionar rótulos e título
ax.set_xlabel('Categorias')
ax.set_ylabel('Ticket Médio (R$)')
ax.set_title('Ticket Médio por Categoria em Todas as Lojas')
ax.set_xticks(x)
ax.set_xticklabels(categorias, rotation=45, ha='right')
ax.legend()

# Exibir o gráfico
plt.tight_layout()
plt.show()
```

# 4. Produtos Mais e Menos Vendidos

```python
# Função para calcular os produtos mais vendidos por loja
def produtos_mais_vendidos_por_loja(df):
    # Contagem de vendas por produto
    contagem_vendas = df['Produto'].value_counts().reset_index()
    contagem_vendas.columns = ['Produto', 'Número de Vendas']
        # Ordenar pelo número de vendas (decrescente)
    contagem_vendas = contagem_vendas.sort_values(by='Número de Vendas', ascending=False)
    return contagem_vendas
```

```python
# Aplicar a função para cada Loja
vendas_por_produto_loja1 = produtos_mais_vendidos_por_loja(loja)
vendas_por_produto_loja2 = produtos_mais_vendidos_por_loja(loja2)
vendas_por_produto_loja3 = produtos_mais_vendidos_por_loja(loja3)
vendas_por_produto_loja4 = produtos_mais_vendidos_por_loja(loja4)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Função para calcular os produtos mais e menos vendidos por loja
def produtos_mais_menos_vendidos(df):
    # Contagem de vendas por produto
    contagem_vendas = df['Produto'].value_counts().reset_index()
    contagem_vendas.columns = ['Produto', 'Número de Vendas']

    # Separar os 5 mais vendidos e os 5 menos vendidos
    mais_vendidos = contagem_vendas.head(5)
    menos_vendidos = contagem_vendas.tail(5).sort_values(by='Número de Vendas', ascending=True)

    return mais_vendidos, menos_vendidos

# Aplicar a função para cada loja
mais_vendidos_loja1, menos_vendidos_loja1 = produtos_mais_menos_vendidos(loja)
mais_vendidos_loja2, menos_vendidos_loja2 = produtos_mais_menos_vendidos(loja2)
mais_vendidos_loja3, menos_vendidos_loja3 = produtos_mais_menos_vendidos(loja3)
mais_vendidos_loja4, menos_vendidos_loja4 = produtos_mais_menos_vendidos(loja4)

# Criar o gráfico
fig, axes = plt.subplots(2, 1, figsize=(12, 10))

# Gráfico dos produtos mais vendidos
x = np.arange(len(mais_vendidos_loja1['Produto']))  # Posições no eixo X
width = 0.2  # Largura das barras

axes[0].bar(x - 1.5 * width, mais_vendidos_loja1['Número de Vendas'], width, label='Loja 1', color='blue')
axes[0].bar(x - 0.5 * width, mais_vendidos_loja2['Número de Vendas'], width, label='Loja 2', color='green')
axes[0].bar(x + 0.5 * width, mais_vendidos_loja3['Número de Vendas'], width, label='Loja 3', color='orange')
axes[0].bar(x + 1.5 * width, mais_vendidos_loja4['Número de Vendas'], width, label='Loja 4', color='red')

axes[0].set_title("Produtos Mais Vendidos por Loja")
axes[0].set_xticks(x)
axes[0].set_xticklabels(mais_vendidos_loja1['Produto'], rotation=45, ha='right')
axes[0].legend()

# Gráfico dos produtos menos vendidos
x = np.arange(len(menos_vendidos_loja1['Produto']))  # Posições no eixo X

axes[1].bar(x - 1.5 * width, menos_vendidos_loja1['Número de Vendas'], width, label='Loja 1', color='blue')
axes[1].bar(x - 0.5 * width, menos_vendidos_loja2['Número de Vendas'], width, label='Loja 2', color='green')
axes[1].bar(x + 0.5 * width, menos_vendidos_loja3['Número de Vendas'], width, label='Loja 3', color='orange')
axes[1].bar(x + 1.5 * width, menos_vendidos_loja4['Número de Vendas'], width, label='Loja 4', color='red')

axes[1].set_title("Produtos Menos Vendidos por Loja")
axes[1].set_xticks(x)
axes[1].set_xticklabels(menos_vendidos_loja1['Produto'], rotation=45, ha='right')
axes[1].legend()

# Ajustar layout
plt.tight_layout()
plt.show()
```
